# project 1
## pseudocode
## request user to input a word
## request user to enter a string of forbidden letters
## if any of the forbidden letter is in the word:
## print "a forbidden letter was used"
## else, return true

```python
# Avoids, the python program.
def avoids():
    word = input("enter a word here            ")
    F = input("enter a forbidden letter        ")
    O = input("enter another forbidden letter  ")
    R = input("enter another forbidden letter  ")
    B = input("enter another forbidden letter  ")
    I = input("enter another forbidden letter  ")
    D = input("enter another forbidden letter  ")
    if 'F' and 'O' and 'R' and 'B' and 'I' and 'D' in word:
        print("A forbidden letter was used")
    else:
        return("True")

avoids()
```
