# pseudocode to get the difference between a given number and 17

![Picture1](images/Picture1.png)

## let B be the given number
## input the value of B
## compute R= B-17
## If B>17:
## D = Rx2
## return D
## Else:
## Q = -R
## return Q



```python
# Python program for project 1
B =int(input("Enter any number of your choice    "))
R = B-17
print("the value of R is", R)
if B>17:
    D = R*2
    print("the double absolute difference is",D)
else:
    Q=-R
    print("the value of Q is",Q)
```

# pseudocode for project 4
## enter the three integer values; a, b and c
## use the max() and min() function to find the largest and smallest intergers respectively
## find the middle using [(a+b+c)-max-min] to get the middle number
## print out the maximun, minimum and middle numbers

```python
# python program for project 4
a = int(input("enter an integer here    "))
b = int(input("enter an integer here    "))
c = int(input("enter an integer here    "))

MAX_NUM = max(a,b,c)
print("the maximum integer is",MAX_NUM)
MIN_NUM = min(a,b,c)
print("the minimum integer is",MIN_NUM)
MID= (a+b+c)-MAX_NUM-MIN_NUM
print("Middle:",MID)
```

```python
# python program for project 5
def cubic_sum(n):
    n -= 1
total = 0
while n > 0:
    total += n*n*n
n -= 1
return total
print(" the sum of cubes smaller than the specified number is",cubic_sum(3))
```
